```python
import numpy as np
import matplotlib.pyplot as plt
# Define the coefficients of the quadratic
a = 1
b = 2
c = 3

# Define the quadratic function
def quadratic_function(x):
  return a*x**2 + b * x + c

x_vertex = -b/(2*a)
y_vertex = quadratic_function(x_vertex)

x_values = np.linspace(x_vertex - 10, x_vertex + 10, 400)
y_values = quadratic_function(x_values)

plt.plot(x_values, y_values , label= f'f(x)={a}x^2 + {b}x + {c}')
plt.scatter(x_vertex, y_vertex , color='red' , label=f'{"minimum" if a> 0 else "maximum"} at ({x_vertex:.2f} , {y_vertex:.2f})')
plt.title("Quadratic Function plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

print (f"the {'minimum' if a>0 else 'maximum' } point of the function is at x = {x_vertex:.2f} , f(x)= {y_vertex:.2f}")


```
